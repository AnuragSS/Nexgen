```python
# Adjust this path if needed
bound_path = Path("CAUTH_MAY_2025_EN_BSC_1271049543973882786.geojson")

bound_gdf = gpd.read_file(bound_path)

print(bound_gdf.head())
print(bound_gdf.columns)
print(bound_gdf.crs)
```

```python
# Make sure CRS is WGS84 (lat/lon) for Folium
if bound_gdf.crs is None or bound_gdf.crs.to_epsg() != 4326:
    bound_gdf = bound_gdf.to_crs(epsg=4326)

# Keep only Liverpool City Region
lcr = bound_gdf[bound_gdf["CAUTH25NM"] == "Liverpool City Region"].copy()

# Get map centre
centre = lcr.geometry.unary_union.centroid
centre_lat, centre_lon = centre.y, centre.x

lcr.head()
```

```python
import folium

m_lcr = folium.Map(
    location=[centre_lat, centre_lon],
    zoom_start=10,
    tiles="OpenStreetMap"
)

folium.GeoJson(
    lcr,
    name="Liverpool City Region",
    style_function=lambda x: {
        "fillColor": "#3186cc",
        "color": "#000000",
        "weight": 2,
        "fillOpacity": 0.2,
    },
    tooltip=folium.GeoJsonTooltip(
        fields=["CAUTH25NM", "CAUTH25CD"],
        aliases=["Region", "Code"]
    )
).add_to(m_lcr)

folium.LayerControl().add_to(m_lcr)

m_lcr
```

```python
# Load brownfield sites
brownfield = gpd.read_file("brownfield-land.geojson")

# Match CRS
if brownfield.crs is None or brownfield.crs.to_epsg() != 4326:
    brownfield = brownfield.to_crs(epsg=4326)

# Keep only sites within Liverpool City Region
brownfield_lcr = gpd.sjoin(
    brownfield,
    lcr[["geometry"]],
    how="inner",
    predicate="within"
)

brownfield_lcr.head(), len(brownfield_lcr)
```

```python
m_bf_lcr = folium.Map(
    location=[centre_lat, centre_lon],
    zoom_start=10,
    tiles="OpenStreetMap"
)

# LCR outline
folium.GeoJson(
    lcr,
    name="Liverpool City Region",
    style_function=lambda x: {
        "fillColor": "#00000000",  # transparent
        "color": "#000000",
        "weight": 2,
        "fillOpacity": 0.0,
    }
).add_to(m_bf_lcr)

# Decide which columns to show in popup if present
popup_candidates = ["SiteName", "local-authority", "hectares", "planning-status"]

for _, row in brownfield_lcr.iterrows():
    geom = row.geometry
    if geom.is_empty:
        continue

    lat = geom.y
    lon = geom.x

    pieces = []
    for col in popup_candidates:
        if col in brownfield_lcr.columns:
            pieces.append(f"{col}: {row.get(col)}")

    popup_text = "<br>".join(pieces) if pieces else None

    folium.CircleMarker(
        location=[lat, lon],
        radius=4,
        color="red",
        fill=True,
        fill_opacity=0.8,
        popup=popup_text,
    ).add_to(m_bf_lcr)

folium.LayerControl().add_to(m_bf_lcr)

m_bf_lcr
```

```python
# Take the Liverpool City Region row directly
lcr_row = lcr.iloc[0]
centre_lat = lcr_row["LAT"]
centre_lon = lcr_row["LONG"]
```

```python
import numpy as np
import branca.colormap as cm
import folium

# Ensure hectares is numeric
brownfield_lcr["hectares"] = pd.to_numeric(brownfield_lcr["hectares"], errors="coerce")

# Simple minmax for size & colour
h_min = brownfield_lcr["hectares"].min()
h_max = brownfield_lcr["hectares"].max()

# Colour scale
cmap = cm.LinearColormap(
    ["#fee5d9", "#fcae91", "#fb6a4a", "#de2d26", "#a50f15"],
    vmin=h_min,
    vmax=h_max
).to_step(5)

m_bf_styled = folium.Map(
    location=[centre_lat, centre_lon],
    zoom_start=10,
    tiles="OpenStreetMap"
)

# LCR outline
folium.GeoJson(
    lcr,
    name="Liverpool City Region",
    style_function=lambda x: {
        "fillColor": "#00000000",
        "color": "#000000",
        "weight": 2,
        "fillOpacity": 0.0,
    }
).add_to(m_bf_styled)

# Add brownfield sites
for _, row in brownfield_lcr.iterrows():
    geom = row.geometry
    if geom.is_empty or pd.isna(row["hectares"]):
        continue

    lat, lon = geom.y, geom.x
    h = row["hectares"]

    # radius scaled 312
    radius = 3 + 9 * (h - h_min) / (h_max - h_min + 1e-9)

    popup_lines = []
    for col in ["name", "reference", "local-authority", "hectares", "planning-permission-status"]:
        if col in brownfield_lcr.columns:
            popup_lines.append(f"{col}: {row.get(col)}")

    folium.CircleMarker(
        location=[lat, lon],
        radius=radius,
        color=cmap(h),
        fill=True,
        fill_opacity=0.85,
        popup="<br>".join(popup_lines) if popup_lines else None,
    ).add_to(m_bf_styled)

# Add legend
cmap.caption = "Site size (hectares)"
cmap.add_to(m_bf_styled)

folium.LayerControl().add_to(m_bf_styled)

m_bf_styled
```

## Understanding the Brownfield Site Popups

Each red dot on the map represents a **brownfield site** within the Liverpool City Region.  
Brownfield sites are previously developed land areas that are now available for redevelopment  often used for future housing, regeneration projects, or mixed-use developments.

When you click on any site, the popup provides a small information summary. Heres what each field means:

---

### ** name: BR074**
This is the sites official **identifier** from the national Brownfield Land Register.  
Its not a place name  its a unique code used by planning authorities to track the site.

---

### ** reference: BR074**
This is the same reference code used across planning documentation.  
If you search this code on the local council planning portal, you can find related applications, reports, or history.

---

### ** hectares: 35.66**
This shows the **total size of the site** in hectares.  
For context:
- **1 hectare  2.47 acres**
- So **35.66 hectares  88 acres**

A site of this size is considered **large and high-capacity**, meaning it could support hundreds of homes or a major mixed-use development.  
In housing analysis, larger sites usually receive **higher feasibility scores**.

---

### ** planning-permission-status: permissioned**
This tells you the **planning status** of the land.

Permissioned means the site already has **approved planning permission**, which is important because:

- It removes major administrative barriers  
- Development can begin sooner  
- It increases the sites feasibility for housing delivery  
- It signals alignment with local authority plans  

In simple terms:  
**A large, permissioned brownfield site is a prime candidate for new housing development.**

---

These popups help highlight *where real development potential exists* by showing the size, status, and readiness of each site.
